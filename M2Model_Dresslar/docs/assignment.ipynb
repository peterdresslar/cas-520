Step 2: Coding
To make the model more realistic, it's unlikely that an unhappy turtle will move to a patch where it knows that it would also be unhappy. Change the unhappy turtle behavior so that instead of moving to a random unoccupied patch, the turtle makes a single move to a patch where it will be happy, and if no such patches exist, the turtle stays put.

Comment your code notating any changes from the original model, for example:

;MODIFIED 3 lines to change the patch color
For this coding exercise, you will likely need a reporter. A netlogo reporter is instructions for computing a value, which then ‘reports’ to whoever asked it. It's like a procedure, but instead of using to we use to-report. Like a procedure, there in also an end. The procedure exits when the report command is used within the reporter, which returns the value of the calculation to the asking entity. A reporter will usually also have an input, which is often some value that is used in the reporter calculation. In general the format for a reporter is

to-report REPORTER-NAME [INPUT]
    LETS AND COMMANDS
    report VALUE
end
The reporter could return numerical values, agentsets, or boolean values. Here's a specific example of a turtle reporter which will return a numeric value :

to-report circle-area [ r ] 
    report ( pi * ( r  ^ 2 ) )
end
This reporter could be called by

ask turtles [ print circle-area size / 2 ]
which will ask each each turtle to print an area value using half its size and the circle-area reporter.  

If a turtle is asking a patch to compute a value using one of the turtle's variables, you will need to use the myself primitive. Here's a boolean patch reporter that compares the turtle size with the patch size. 

to-report compare-size [ turtle-size ] 
     ifelse turtle-size = patch-size
          [report true]
          [report false]
end

ask turtles [ 
    ask patches [ 
        ifelse compare-size [ size ] of myself
            [print 'yes']
            [print 'no' ]
    ]
]
*Please note that it is possible to complete this assignment without a reporter. You will not be penalized if you decide to complete this assignment in that way.
